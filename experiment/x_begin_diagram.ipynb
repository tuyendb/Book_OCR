```python
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import json
import numpy as np

with open('../data4.json') as jsonfile:
    data = json.load(jsonfile)
    paragr_x_begin = []
    for id in data.keys():
        paragr_x_begin.append(data[id]['x_begin'])
new_paragr_x_begin = sorted(list(set(paragr_x_begin)))
paragr_x_begin_amount = []
for x_begin in new_paragr_x_begin:
    paragr_x_begin_amount.append(paragr_x_begin.count(x_begin))
max_x_begin = max(new_paragr_x_begin)
compare_x_begin = [f for f in range(max_x_begin + 1)]
new_paragr_x_begin_amount = []
for cx in compare_x_begin:
    if cx in new_paragr_x_begin:
        id = new_paragr_x_begin.index(cx)
        new_paragr_x_begin_amount.append(paragr_x_begin_amount[id])
    else:
        new_paragr_x_begin_amount.append(0)
new_paragr_x_begin_amount = np.array(new_paragr_x_begin_amount)
peaks, p_probs = find_peaks(new_paragr_x_begin_amount, height=10)
troughs, t_probs = find_peaks(-new_paragr_x_begin_amount)
plt.plot(new_paragr_x_begin_amount)
plt.xlabel("x_begin")
plt.ylabel("Amount")
plt.plot(peaks, new_paragr_x_begin_amount[peaks], '^')
plt.show()
print(peaks)
```
